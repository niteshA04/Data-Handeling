# Handeling Mixed Variables

### Importing Libraries

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.sample(10)
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling wih')
```

### Extract Numerical Part

```python
df['number_n'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
df.sample(10)
```

### Extract Categorical Part

```python
df['number_c'] = np.where(df['number_n'].isnull(),df['number'],np.nan)
df.sample(10)
```
